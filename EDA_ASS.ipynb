```python
'''● Compare average watch_hours and monthly_fee across:
○ subscription_type
○ region
○ device
● Plot average avg_watch_time_per_day by favorite_genre.
● Analyze churn rate based on:
○ gender
○ region
○ subscription_type
○ payment_method'''
```

```python
# subscription type
```

```python
avg_data = df.groupby('subscription_type')[['watch_hours', 'monthly_fee']].mean()
avg_data
```

```python
sns.barplot(x=avg_data.index, y=avg_data['watch_hours'])
plt.title("Average Watch Hours by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Average Watch Hours")
plt.show()
```

```python
sns.barplot(x=avg_data.index, y=avg_data['monthly_fee'])
plt.title("Average Monthly Fee by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Average Monthly Fee (₹)")
plt.show()
```

```python
# Region 
```

```python
avg1 = df.groupby('region')[['watch_hours', 'monthly_fee']].mean()
avg1
```

```python
sns.barplot(x=avg1.index, y=avg1['watch_hours'])
plt.title("Average Watch Hours by Region")
plt.xlabel("Region")
plt.ylabel("Average Watch Hours")
plt.show()
```

```python
sns.barplot(x=avg1.index, y=avg1['monthly_fee'])
plt.title("Average Monthly Fee by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Average Monthly Fee (₹)")
plt.show()
```

```python
# device
```

```python
avg2 = df.groupby('device')[['watch_hours', 'monthly_fee']].mean()
avg2
```

```python
sns.barplot(x=avg2.index, y=avg2['watch_hours'])
plt.title("Average Watch Hours by device")
plt.xlabel("device")
plt.ylabel("Average Watch Hours")
plt.show()
```

```python
sns.barplot(x=avg2.index, y=avg2['monthly_fee'])
plt.title("Average Monthly Fee by cevice")
plt.xlabel("device")
plt.ylabel("Average Monthly Fee (₹)")
plt.show()
```

```python
#Plot average avg_watch_time_per_day by favorite_genre.
```

```python
avg_watch = df.groupby('favorite_genre')['avg_watch_time_per_day'].mean().reset_index()
```

```python
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=avg_watch)
plt.title("Average Watch Time per Day by Favorite Genre")
plt.xlabel("Favorite Genre")
plt.ylabel("Avg Watch Time per Day (hours)")
plt.show()
```

```python
# watch_hours is negative predicator of churned
```

```python
# Premium users vhurn less than basic users
# Despite higher cost, Premium users are more loyal, suggesting that investing in higher-tier plans might improve retention.
# Basic plan may lack features or value, leading to user dissatisfaction and eventual churn.
# Standard and Premium users have more balanced churn vs non-churn proportions, while Basic users have much higher churn volume than retention.
# Targeted retention efforts like offering feature trials or personalized offers to Basic users could significantly reduce churn.
```

```python
# On the basis of average watch hours by region North America has the highest average
# South America and Oceania follow closely, suggesting strong user engagement in these regions as well.
# Asia has the lowest average watch hours (≈ 11.28), which could indicate lower engagement or content relevance in that region.
# The difference in average watch time across regions is not drastic, but still notable enough to guide content and marketing strategies.
# Regions with higher average watch hours may respond better to content recommendations and upselling offers.
```
